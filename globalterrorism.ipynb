# STEP:3 Creating New Data Frame

```python
data.rename(columns={'iyear':'YEAR','imonth':'MONTH','iday':'DAY','country_txt':'COUNTRY','region_txt':'REGION','provstate':'State','attacktype1_txt':'ATTACK-TYPE',
                     'targtype1_txt':'TARGET','nkill':'Kill','city':'CITY','targsubtype1_txt':'DEPARTMENT','motive':'MOTIVE','weaptype1_txt':'Weapon_type','longitude':'Longitude','latitude':'Latitude',},inplace =True)
```

```python
df=data[['YEAR','MONTH','DAY','COUNTRY','REGION','State','CITY','ATTACK-TYPE','Kill','TARGET','DEPARTMENT','MOTIVE','Weapon_type','Longitude','Latitude']]
```

```python
df.head()
```

```python
df.shape
```

# Step:-4 Cleaning the data set

```python
df.isnull().sum()
```

```python
df.fillna('None',inplace=True)
```

```python
df.isnull().sum()
```

# Step:-5 Analyze Attack-type and country column

```python
## Find the different attack in attack-type column in data-set df.
df['ATTACK-TYPE'].unique()
```

```python
## Here we find total number of different country where attack happens. 
len(df['COUNTRY'].unique())
```

```python
### Here we plot bar graph of attack-type on the basis of country.
df.groupby('ATTACK-TYPE').count()['COUNTRY'].sort_values(ascending=False).plot(kind='bar')
```

####
From above graph we clearly see most of the attck happen by Bombing/Explosion in country.

```python
### Here we plot bar graph of country on the basis of attack-type.
df.groupby('COUNTRY').count()['ATTACK-TYPE'].nlargest(14).sort_values(ascending=False).plot(kind='bar')
```

####
From above graph we clearly see that most of the attack happen in iraq then in pakistan,afghanistan,India,Colombia in decreasing order.
